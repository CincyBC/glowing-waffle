```python
f = open('slopes.txt', 'r')
slopes = f.read().split('\n')
print(slopes)
```

```python
full_slopes = []
multiplier = (len(slopes) // len(slopes[0])) + 1
multiplier *= 7
for x in slopes:
    new = x * multiplier
    full_slopes.append(new)
```

```python
trees_second = 0
for x in range(len(full_slopes)):
    if x == 0:
        pass
    multi = x * 3
    row = full_slopes[x]
    if row[multi] == "#":
        trees_second += 1
print(trees_second)   
```

```python
"""
Right 1, down 1.
Right 3, down 1. (This is the slope you already checked.)
Right 5, down 1.
Right 7, down 1.
Right 1, down 2.
"""
```

```python
trees_first = 0
for x in range(len(full_slopes)):
    if x == 0:
        pass
    multi = x 
    row = full_slopes[x]
    if row[multi] == "#":
        trees_first += 1
print(trees_first)   
```

```python
trees_third = 0
for x in range(len(full_slopes)):
    if x == 0:
        pass
    multi = x * 5
    row = full_slopes[x]
    if row[multi] == "#":
        trees_third += 1
print(trees_third)   
```

```python
trees_fourth = 0
for x in range(len(full_slopes)):
    if x == 0:
        pass
    multi = x * 7
    row = full_slopes[x]
    if row[multi] == "#":
        trees_fourth += 1
print(trees_fourth) 
```

```python
trees_fifth = 0
for x in range(len(full_slopes) // 2):
    if x == 0:
        pass
    else:
        multi = x
        row = full_slopes[x * 2]
        if row[multi] == "#":
            trees_fifth += 1
print(trees_fifth) 
```

```python
print(trees_first, trees_second, trees_third, trees_fourth, trees_fifth)
print(trees_first * trees_second * trees_third * trees_fourth * trees_fifth)
```
