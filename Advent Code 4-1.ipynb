```python
import re
f = open('passports.txt', 'r')
passports = f.read().split('\n\n')
print(passports)
```

```python
print(len(passports))
```

```python
required = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']
valid = []
for x in passports:
    numFalse = 0
    for z in required:
        if z + ':' in x:
            pass
        else:
            numFalse += 1
    if numFalse == 0:
        valid.append(x)

print(len(valid))
```

```python
"""
byr (Birth Year) - four digits; at least 1920 and at most 2002.
iyr (Issue Year) - four digits; at least 2010 and at most 2020.
eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
hgt (Height) - a number followed by either cm or in:
If cm, the number must be at least 150 and at most 193.
If in, the number must be at least 59 and at most 76.
hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
pid (Passport ID) - a nine-digit number, including leading zeroes.
cid (Country ID) - ignored, missing or not.
"""
```

```python
# byr (Birth Year) - four digits; at least 1920 and at most 2002.
def byr_validator(x):
    first = x.index('byr:') + 4
    end = first + 4
    year = x[first:end]
    if int(year) > 1919 and int(year) < 2003:
        return 0
    else:
        return 1
```

```python
# iyr (Issue Year) - four digits; at least 2010 and at most 2020.
def iyr_validator(x):
    first = x.index('iyr:') + 4
    end = first + 4
    year = x[first:end]
    if int(year) > 2009 and int(year) < 2021:
        return 0
    else:
        return 1
```

```python
# eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
def eyr_validator(x):
    first = x.index('eyr:') + 4
    end = first + 4
    year = x[first:end]
    if int(year) > 2019 and int(year) < 2031:
        return 0
    else:
        return 1
```

```python
# pid (Passport ID) - a nine-digit number, including leading zeroes.
def pid_validator(x):
    first = x.index('pid:') + 4
    end = first + 9
    pidno = x[first:end]
    if pidno.isdigit() == True:
        if int(pidno) < 9:
            return 1
        else:
            return 0
    else:
        return 1
```

```python
# hgt (Height) - a number followed by either cm or in:
# If cm, the number must be at least 150 and at most 193.
# If in, the number must be at least 59 and at most 76.
def hgt_validator(x):
    first = x.index('hgt:') + 4
    end = first + 7
    possible_answer = x[first:end]
    if "in" in possible_answer or "cm" in possible_answer:
        if "in" in possible_answer:
            inches = int(possible_answer.split('in')[0])
            if inches < 59 or inches > 76:
                print(inches, "false")
                return 1
            else:
                print(inches, "true")
                return 0
        if "cm" in possible_answer:
            cents = int(possible_answer.split('cm')[0])
            if cents < 150 or cents > 193:
                print(cents, "false")
                return 1
            else:
                print(cents, "true")
                return 0
    else:
        print(possible_answer, 'total failure')
        return 1
```

```python
# hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
def hcl_validator(x):
    first = x.index('hcl:') + 4
    end = first + 7
    string = x[first:end]
    match = re.search(r'^#(?:[0-9a-fA-F]{3}){1,2}$', string)
    if match:
        return 0
    else:
        return 1
```

```python
# ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
def ecl_validator(x):
    first = x.index('ecl:') + 4
    validColor = ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    end = first + 3
    color = x[first:end]
    if color in validColor:
        return 0
    else:
        return 1
```

```python
valids = 0
for x in valid:
    results = 0
    results += byr_validator(x)
    results += iyr_validator(x)
    results += eyr_validator(x)
    results += pid_validator(x)
    results += hgt_validator(x)
    results += ecl_validator(x)
    results += hcl_validator(x)            
                
    if results < 1:
        valids += 1

print(valids, ' of ', len(valid))
```
