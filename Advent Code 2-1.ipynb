```python
f = open('password_list.txt', 'r')
```

```python
passwords = f.read()
passwords = passwords.split('\n')
print(passwords)
```

```python
#Break everything up into their useable parts
master_list = []
for x in passwords:
    temp = x.split(':')
    rules = temp[0].split(' ')
    numbers = rules[0].split('-')
    first = numbers[0]
    second = numbers[1]
    letter = rules[1]
    password = temp[1].lstrip()
    master_list.append([int(first), int(second), letter, password])
    print([first, second, letter, password])
```

```python
#Get the ones that follow rules
print(len(master_list))
correct = 0
for x in master_list:
    string = x[3]
    counts = string.count(x[2])
    if x[0] <= counts and x[1] >= counts:
        correct += 1
        
print(correct)
    
```

```python
"""
Each policy actually describes two positions in the password, where 1 means the first character, 
2 means the second character, and so on. (Be careful; Toboggan Corporate Policies have no concept of 
"index zero"!) Exactly one of these positions must contain the given letter. Other occurrences of the 
letter are irrelevant for the purposes of policy enforcement.

Given the same example list from above:

1-3 a: abcde is valid: position 1 contains a and position 3 does not.
1-3 b: cdefg is invalid: neither position 1 nor position 3 contains b.
2-9 c: ccccccccc is invalid: both position 2 and position 9 contain c.
"""
```

```python
second_correct = 0
for x in master_list:
    tempNum = 0
    string = x[3]
    first = x[0] - 1
    second = x[1] - 1
    if string[first] == x[2]:
        tempNum += 1
    if string[second] == x[2]:
        tempNum += 1
    if tempNum == 1:
        second_correct += 1
        
print(second_correct)
```
