```python
#Break up list of numbers and convert to integers
numbers = numbers.split(' ')
numbers = [int(x) for x in numbers]
print(numbers)
```

```python
#If X + Y = 2020, then either X or Y must be smaller than 1010, so divide the numbers for faster iteration
smaller = []
larger = []
for x in numbers:
    if x <= 1010:
       smaller.append(x) 
    else:
        larger.append(x)
print(smaller, larger)
```

```python
# Find the winning pair by adding each from the smaller group to each one in the larger group
winning_pair = []
for L in larger:
    for x in smaller:
        if x + L == 2020:
            winning_pair.append(x)
            winning_pair.append(L)
            
print(winning_pair)
```

```python
print(winning_pair[0] * winning_pair[1])
```

```python
#Now we have to find 3 numbers that = 2020.
smaller.sort()
print(smaller, smaller[0] + smaller[1])
print(2020 - smaller[0] - smaller[1])
```

```python
three_set = []
for x in numbers:
    if x <= 1442:
       three_set.append(x) 
winning_three = []
for x in range(len(three_set)):
    y = x + 1
    z = x + 2
    while y < len(three_set) - 1:
        y += 1
        while z < len(three_set) - 2:
            z += 1
            if three_set[x] + three_set[y] + three_set[z] == 2020:
                winning_three.append(three_set[x])
                winning_three.append(three_set[y])
                winning_three.append(three_set[z])
    
    
```
